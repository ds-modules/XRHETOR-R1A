## Part 2: Segmented analysis

**When predicting election outcomes, its important to look for certain segments within a population that strongly favor one candidate over the other**

Take a look at the poll data. For each cell, take a look at the code, run it (shift + enter) then write down any observations you have for that chart.

## First, lets look to see if gender influences political preferences

```python
#read in the data
pollData = Table().read_table("Poll_Data.csv")

genderPollData = pollData.select(["Men", "Women"])
genderPollData.barh(["% for Clinton", "% for Trump"])
```

Observations:

## Lets also look at how political opinions vary between 3 prominent racial groups

```python
ethnicityPollData = pollData.select(["White", "Black", "Hispanic"])
ethnicityPollData.barh(["% for Clinton", "% for Trump"])
```

Observations:

## Other demographics can be helpful as well, lets see how age influences preference below

```python
agePollData = pollData.select(["18 to 29", "30 to 49", "50 to 64", "65 and older"])
agePollData.barh(["% for Clinton", "% for Trump"])
```

Observations: 

## Education level can also show some stark differences, what do you think would cause this?

```python
eduPollData = pollData.select(["Postgrad", "College Grad", "Some College", "HS or Less"])
eduPollData.barh(["% for Clinton", "% for Trump"])
```

Observations: 

## Model Assessment

**Its important to remember that no single indicator is perfect and that many variables will influence voter preferences, for example, you can see that people even defy party lines about *1 in 20* times.**

```python
partyPollData = pollData.select(["Rep or Lean Rep", "Dem or Lean Dem"])
partyPollData.barh(["% for Clinton", "% for Trump"])
```
