```python
import pandas as pd
import numpy as np
import json
```

```python
haidt = pd.read_csv("raw_dict.csv", header=None)
```

```python
haidt.head()
```

```python
care = (haidt[0][haidt[0].str.contains(r'(01)|(02)')]
        .str.replace(r'[^A-Za-z]', '')
        .tolist())
fair = (haidt[0][haidt[0].str.contains(r'(03)|(04)')]
        .str.replace(r'[^A-Za-z]', '')
        .tolist())
loyal = (haidt[0][haidt[0].str.contains(r'(05)|(06)')]
         .str.replace(r'[^A-Za-z]', '')
        .tolist())
auth = (haidt[0][haidt[0].str.contains(r'(07)|(08)')]
        .str.replace(r'[^A-Za-z]', '')
       .tolist())
pure = (haidt[0][haidt[0].str.contains(r'(09)|(10)')]
        .str.replace(r'[^A-Za-z]', '')
       .tolist())
```

```python
haidt_dict = {'care/harm': care,
             'fairness/cheating': fair,
             'authority/subversion': auth,
             'loyalty/betrayal': loyal,
             'sanctity/degradation': pure}
haidt_dict
```

```python
with open('../mft_data/haidt_dict.json', 'w') as fp:
    json.dump(haidt_dict, fp, sort_keys=True, indent=4)
```
